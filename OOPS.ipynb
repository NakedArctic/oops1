Q(1)What is Object-Oriented Programming (OOP)?
- Object-Oriented Programming (OOP)
Object-Oriented Programming (OOP) is a programming paradigm that organizes software design around objects rather than functions or logic. Objects are instances of classes that encapsulate data (attributes) and behavior (methods). OOP helps in writing reusable, scalable, and modular code.

Key Principles of OOP
Encapsulation ‚Äì Bundling data and methods that operate on the data within a single unit (class).
Abstraction ‚Äì Hiding complex implementation details and exposing only the necessary functionalities.
Inheritance ‚Äì Enabling a new class to inherit properties and behavior from an existing class.
Polymorphism ‚Äì Allowing a single interface to represent different types or enabling the same method to work in different ways.

Q(4)What is the difference between abstraction and encapsulation?
- Difference Between Abstraction and Encapsulation in OOP
Both abstraction and encapsulation are fundamental principles of Object-Oriented Programming (OOP), but they serve different purposes.

1. Abstraction
Definition:
Abstraction is the process of hiding unnecessary details and showing only the essential features of an object.

Purpose:
To reduce complexity and increase efficiency by exposing only relevant details to the user.

Example:
When you drive a car, you use the steering wheel and pedals without needing to know how the engine works.

In Python (Abstraction using Abstract Classes)
python
Copy
Edit
from abc import ABC, abstractmethod

class Animal(ABC):  # Abstract Class
    @abstractmethod
    def make_sound(self):  # Abstract Method
        pass

class Dog(Animal):
    def make_sound(self):
        return "Woof!"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"

- Creating objects
dog = Dog()
print(dog.make_sound())  # Output: Woof!
üîπ Why is this abstraction?

The Animal class provides a general concept of an animal but does not define how the make_sound() method works.
The details are implemented in the subclasses (Dog, Cat).
2. Encapsulation
Definition:
Encapsulation is the process of wrapping data and methods into a single unit (class) and restricting direct access to some of the object's details.

Purpose:
To protect data from unintended modification by restricting direct access using access modifiers.

Example:
A bank account hides the balance variable and only allows modification through methods.

In Python (Encapsulation using Private Variables)
python
Copy
Edit
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited ${amount}, New Balance: ${self.__balance}"
        return "Invalid amount"

    def get_balance(self):
        return f"Current Balance: ${self.__balance}"

- Creating an object
account = BankAccount(1000)

- Accessing balance via method (Encapsulation)
print(account.get_balance())  # Output: Current Balance: $1000

- Trying to access private attribute directly (will cause error)
- print(account.__balance)  # AttributeError
üîπ Why is this encapsulation?

The balance is private (__balance) and cannot be accessed directly.
Instead, we use getter (get_balance()) and setter (deposit()) methods to safely interact with the data.
Key Differences:
Feature	Abstraction	Encapsulation
Definition	Hiding complex implementation details	Restricting direct access to data & methods
Focus On	Hiding implementation	Hiding data
Purpose	Expose only necessary details	Protect and control access to data
Achieved Through	Abstract classes and interfaces	Private/protected attributes and methods
Example	Steering a car without knowing engine details	A bank account hiding balance from direct access

- Dunder Methods in Python (Magic Methods)
Dunder methods (short for "double underscore methods") are special methods in Python that have double underscores (__) at the beginning and end of their names. They are also called magic methods because they enable built-in behaviors for objects, such as initialization, string representation, addition, and comparison.

1. Commonly Used Dunder Methods
Here are some of the most important dunder methods with examples:

1.1 __init__ (Constructor)
Called when a new instance of a class is created (like a constructor).

python
Copy
Edit
class Person:
    def __init__(self, name, age):  # Initialization
        self.name = name
        self.age = age

p = Person("Alice", 25)  # Calls __init__()
print(p.name)  # Output: Alice
1.2 __str__ (String Representation)
Defines a human-readable string representation of an object.

python
Copy
Edit
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):  # Called by str() or print()
        return f"{self.name}, {self.age} years old"

p = Person("Alice", 25)
print(p)  # Output: Alice, 25 years old
üîπ Without __str__, print(p) would return an unreadable default string like <__main__.Person object at 0x...>

Used for debugging; should return a string that can recreate the object.

Q(6)Explain the concept of inheritance in OOP?
- Inheritance in Object-Oriented Programming (OOP)
Definition:
Inheritance is an OOP concept where a child class (or subclass) inherits the properties and methods of a parent class (or superclass). It allows code reuse, hierarchy creation, and the extension of existing functionalities without modifying the original class.

1. Why Use Inheritance?
‚úÖ Code Reusability ‚Äì Avoid rewriting code by reusing existing logic.
‚úÖ Hierarchy & Organization ‚Äì Establish a clear relationship between classes.
‚úÖ Extensibility ‚Äì Modify or extend existing functionality in a structured way.
‚úÖ Polymorphism Support ‚Äì Allows overriding methods for dynamic behavior.

Q(7) What is polymorphism in OOP?
-
Polymorphism in Object-Oriented Programming (OOP)
Definition:
Polymorphism means "many forms" and allows different objects to be treated as instances of the same class, even if they behave differently. It enables the same interface (method or operator) to be used with different types of objects.

1. Why Use Polymorphism?
‚úÖ Code Flexibility ‚Äì Write more generic and reusable code.
‚úÖ Simplified Code Maintenance ‚Äì Reduce conditional statements (if-else).
‚úÖ Extensibility ‚Äì Easily add new functionality without modifying existing code.
‚úÖ Supports OOP Principles ‚Äì Works well with inheritance and abstraction.

2. Types of Polymorphism in Python
2.1 Method Overriding (Runtime Polymorphism)
Definition: A child class provides a different implementation for a method already defined in the parent class.
Happens at runtime.
Example:
python
Copy
Edit

for animal in animals:
    print(animal.make_sound())

- Output
- Woof
- Meow
- Some generic sound
üîπ Why is this polymorphism?

The make_sound() method behaves differently based on the actual object type.
The same method name (make_sound) is used in different classes with different behaviors.
2.2 Method Overloading (Compile-time Polymorphism)
üîπ Python does NOT support true method overloading (like Java or C++), but we can achieve similar behavior using default arguments or variable arguments (*args, **kwargs).

Example Using Default Arguments:

Copy
Edit

Q(8)How is encapsulation achieved in Python?
- Encapsulation in Python (OOP)
Definition:
Encapsulation is the bundling of data (variables) and methods that operate on that data into a single unit (class) while restricting direct access to some details. This helps in data hiding and protection.

1. How is Encapsulation Achieved in Python?
Python uses access modifiers (public, protected, and private) to control access to variables and methods.

1.1 Public Members (No Underscore)
Can be accessed from anywhere.
No special syntax is required.

Edit

üîπ Public members are accessible from outside the class, making them not encapsulated.

1.2 Protected Members (Single Underscore _var)
Can be accessed inside the class and its subclasses.
Indicated by a single underscore _.
Not fully private (Python doesn‚Äôt enforce strict protection).

Copy
Edit

    def __init__(self, name, age):

print(emp._name)      # ‚ö†Ô∏è Allowed but not recommended
üîπ Protected members should be accessed only within the class or subclass but are not enforced by Python.

1.3 Private Members (Double Underscore __var)
Cannot be accessed directly outside the class.
Indicated by double underscore __.
Python name-mangles private variables (_ClassName__var).
python
Copy
Edit
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private attribute

        return self.__balance  # Accessible inside the class

üîπ Private members are meant to be accessed only inside the class.
üîπ Python does not strictly enforce privacy but uses name mangling.

2. Using Getters and Setters (Controlled Access)
Encapsulation ensures data safety by using getter and setter methods instead of directly modifying attributes.

python
Copy
Edit
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private attribute

    def get_balance(self):  # Getter method
        return self.__balance

    def set_balance(self, amount):  # Setter method
        if amount >= 0:
            self.__balance = amount
        else:
            print("Invalid balance amount!")

account = BankAccount(1000)

account.set_balance(2000)     # ‚úÖ Updating balance
print(account.get_balance())  # ‚úÖ Output: 2000
account.set_balance(-500)     # ‚ùå Invalid balance amount!
üîπ Encapsulation ensures:
‚úÖ Data protection by making attributes private.
‚úÖ Controlled modification using getter and setter methods.

Instead of using get_ and set_ methods, Python allows property decorators (@property) to create getter and setter functions.

python
Copy
Edit

account.balance = 2000   # ‚úÖ Using setter
print(account.balance)   # ‚úÖ Output: 2000
account.balance = -500   # ‚ùå Invalid amount!
üîπ Advantages of @property Decorators:
‚úÖ More Pythonic & Readable (No explicit get_ or set_ methods).
‚úÖ Encapsulation with Simplicity.

Python provides two special types of methods that are different from regular instance methods:

1. Class Methods (@classmethod)
Definition:
A class method operates on the class level rather than on an instance.
It is defined using the @classmethod decorator.
The first parameter is cls, which represents the class itself, not the instance.
1.1 Syntax & Example

- ‚úÖ Output: Accessing: I am a class variable
üîπ Key Points:
‚úÖ Class methods can access and modify class variables (cls.var).
‚úÖ They do not operate on instance attributes (self.var).

        self.name = name  # Instance variable

        return f"Name: {self.name}, School: {self.school}"

Q(11) What is method overloading in Python?
- Method Overloading in Python
Definition:
Method Overloading refers to the ability to define multiple methods with the same name but different parameters (i.e., the method signature changes).
However, Python does not support traditional method overloading as seen in languages like Java or C++. In Python, if you define a method multiple times, the last definition will override the previous ones.
1. Why is Method Overloading Not Supported Directly in Python?
Python allows dynamic typing, so the same method name can take different types of arguments, but it doesn't enforce a strict signature like Java or C++.
If you define a method multiple times, Python just replaces the older version with the new one.
Example: Method Overloading in Other Languages
In languages like Java, you can have multiple methods with the same name but different parameter types:

java
Copy
Edit
public void display(String name) {
    System.out.println("Hello, " + name);
}

public void display(int age) {
    System.out.println("Age: " + age);
}
But in Python, if you define a method with the same name multiple times, it simply overwrites the previous method.

2. Simulating Method Overloading in Python
Since Python doesn't support multiple method definitions with different parameters directly, you can simulate method overloading using:

Default Arguments
Variable-Length Arguments (*args, **kwargs)
2.1 Using Default Arguments
You can simulate overloading by setting default values for method parameters.

Example: Using Default Arguments
python
Copy
Edit
class Example:
    def display(self, name="John", age=None):
        if age:
            print(f"Name: {name}, Age: {age}")
        else:
            print(f"Name: {name}")

- Calling with different arguments
e = Example()
e.display("Alice", 25)  # ‚úÖ Output: Name: Alice, Age: 25
e.display("Bob")         # ‚úÖ Output: Name: Bob
üîπ Here: The age parameter is optional, and if not passed, it defaults to None.

2.2 Using Variable-Length Arguments (*args, **kwargs)
You can use *args (for non-keyword arguments) and **kwargs (for keyword arguments) to allow a method to accept a variable number of arguments.

Example: Using *args for Overloading Simulation

Copy
Edit

        return sum(args)

print(calc.add(3, 5, 7, 2))  # ‚úÖ Output: 17
üîπ Here: The add method can take any number of arguments, and it calculates the sum, simulating the behavior of an overloaded method with multiple signatures.

üîπ Here: The hello method can handle different sets of arguments (using **kwargs), simulating overloading with different parameter counts.

Q(12)What is method overriding in OOP?
- Method Overriding in OOP (Object-Oriented Programming)
Definition:
Method Overriding occurs when a child class provides its own implementation of a method that is already defined in its parent class.
The child class's method replaces the parent class method when called on an instance of the child class.
It allows the child class to modify or extend the behavior of the inherited method.
1. How Method Overriding Works
The child class inherits the method from the parent class.
The child can then override this inherited method to change its functionality.
When the overridden method is called, the child class's version of the method is executed, not the parent‚Äôs.
2. Syntax of Method Overriding in Python
python
Copy
Edit
class Parent:
    def greet(self):
        print("Hello from Parent!")

class Child(Parent):
    def greet(self):  # Overriding Parent's greet method
        print("Hello from Child!")

- Creating instances
parent = Parent()
child = Child()

parent.greet()  # ‚úÖ Output: Hello from Parent!
child.greet()   # ‚úÖ Output: Hello from Child!
üîπ Key Points:

The child class defines a method greet() with the same name as in the parent class.
When called on the child object, the child's version of the method is executed, overriding the parent‚Äôs version.
3. Purpose of Method Overriding
Change or extend behavior: Method overriding allows child classes to modify or extend the functionality of inherited methods from the parent class.
Dynamic method resolution: The correct version of the method is determined at runtime, allowing for polymorphism.
Improved flexibility: It provides more control over the inherited behavior and enables creating more specialized classes.
4. Example: Overriding a Method in Child Class
python
Copy
Edit
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):  # Overriding the parent method
        print("Woof! Woof!")

class Cat(Animal):
    def speak(self):  # Overriding the parent method
        print("Meow! Meow!")

- Creating objects
dog = Dog()
cat = Cat()

dog.speak()  # ‚úÖ Output: Woof! Woof!
cat.speak()  # ‚úÖ Output: Meow! Meow!
üîπ In this example, both the Dog and Cat classes override the speak() method from the Animal class, each providing its own version of the method.

5. Method Overriding and Polymorphism
Polymorphism allows you to call the same method on different objects (instances of different classes), and the appropriate method for each object type will be executed.
Example of Polymorphism through Method Overriding

                    # Woof! Woof!  (Dog)

üîπ Polymorphism is achieved as we call the same speak() method on different objects (instances of Dog, Cat, and Bird), and each one calls its overridden method.

6. Method Overriding Rules
Same Method Signature: The method in the child class must have the same name and same parameters as the method in the parent class.
Accessing Parent's Method: You can still call the parent class‚Äôs method from the child class using super(). This is useful if you want to extend the parent‚Äôs functionality, rather than completely replacing it.

7. Restrictions on Method Overriding
Method overriding is only allowed for instance methods. It doesn‚Äôt apply to static methods or class methods, as these are bound to the class and not the instance.
You cannot override a method if you are not overriding the same method signature (same name and parameters).
8. Method Overriding vs. Method Overloading
Method Overriding: When a child class provides a specific implementation of a method that is already defined in the parent class. It requires the same method signature.
Method Overloading: Not directly supported in Python, but can be simulated using default arguments or *args/**kwargs to define multiple behaviors for the same method name.

Q(13)What is a property decorator in Python?
-Property Decorator in Python
Definition:
The @property decorator in Python is used to define a method that can be accessed like an attribute, without explicitly calling it as a method.
It allows you to encapsulate data and provide custom getter logic while still accessing it using attribute-style syntax (i.e., without parentheses).
1. Why Use the @property Decorator?
Encapsulation: You can control how an attribute is accessed or modified.
Custom logic: It allows you to add getter, setter, and deleter methods for an attribute, while still presenting the attribute like a regular field.
Read-only properties: You can create read-only attributes that can only be accessed (but not modified).
2. Basic Syntax of @property
python
Copy
Edit
class MyClass:
    def __init__(self, value):
        self._value = value  # Private attribute

    @property
    def value(self):
        # Getter method
        return self._value

- Creating an instance
obj = MyClass(10)
print(obj.value)  # ‚úÖ Output: 10
üîπ Key Points:

@property allows the value() method to be accessed as an attribute, so you can access obj.value instead of calling it as a method (obj.value()).
The value method is now a getter for the _value attribute.
3. Read-only Property (Getter)
The getter function is defined using the @property decorator. This provides access to the private attribute, while hiding the actual implementation.

Example: Read-only Property
python
Copy
Edit
class Circle:
    def __init__(self, radius):
        self._radius = radius  # Private attribute

    @property
    def radius(self):
        return self._radius  # Getter method

    @property
    def area(self):
        return 3.14 * (self._radius ** 2)  # Compute area based on radius

# Creating an instance
circle = Circle(5)

print(circle.radius)  # ‚úÖ Output: 5
print(circle.area)    # ‚úÖ Output: 78.5
üîπ Here, radius is a read-only property, and area is a computed property based on the radius. You don't need to explicitly call these methods.

4. Setter and Getter Methods with @property
You can also use the @property decorator along with @<property_name>.setter to define setters (for modifying values), allowing you to validate or modify data when setting a property.

Example: Getter and Setter
python
Copy
Edit
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    @property
    def name(self):

        if len(value) < 3:

    def age(self):
        return self._age

    @age.setter

            raise ValueError("Age cannot be negative")

- person.name = "Al"  # ‚ùå Raises ValueError: Name must be at least 3 characters long
üîπ In this example, name and age are properties with both getter and setter methods:

The getter (@property) provides access to the private attributes.
The setter (@<property_name>.setter) validates and updates the attributes.
5. Deleter Method
You can also define a deleter method using the @<property_name>.deleter decorator, which allows you to delete a property in a controlled way.

Edit

üîπ Deleter method provides control over property deletion. In this case, when del product.price is called, the price attribute is deleted, and a custom message is displayed.

Q(14) Why is polymorphism important in OOP?
- Why Polymorphism is Important in OOP
Definition:
Polymorphism in Object-Oriented Programming (OOP) refers to the ability of different classes to respond to the same method or function call in a way that is appropriate for their type.
In simple terms, polymorphism allows one interface (method name) to be used for different underlying forms (different classes or objects).
1. Key Benefits of Polymorphism
1.1. Code Reusability
Polymorphism allows for writing more generic code that works with objects of different types. Instead of writing different versions of the same function for each class, you can write one method and allow each class to handle it in its own way.
This leads to reusable code and less duplication.
Example: Code Reusability
python

class Dog:

        return "Woof"

class Cat:
    def speak(self):
        return "Meow"

class Bird:
    def speak(self):
        return "Chirp"

def animal_sound(animal):
    print(animal.speak())

- Creating instances
dog = Dog()
cat = Cat()
bird = Bird()

- Using the same function for different objects
animal_sound(dog)   # ‚úÖ Output: Woof
animal_sound(cat)   # ‚úÖ Output: Meow
animal_sound(bird)  # ‚úÖ Output: Chirp
üîπ Here: animal_sound() works for all types of animals (Dog, Cat, Bird), despite them having different speak() methods. Polymorphism allows this generic approach.

1.2. Flexibility and Maintainability
With polymorphism, you can modify or extend the behavior of your program without altering the interface. If a new class is added, you don‚Äôt need to modify the existing code that uses the common interface. This makes your system more flexible and easier to maintain.
Example: Adding New Behavior
python

- Now you can pass a Cow object to the animal_sound function without changing it

animal_sound(cow)   # ‚úÖ Output: Moo
üîπ Here: Even after adding the Cow class, the animal_sound() function still works without any changes, which demonstrates the flexibility of polymorphism.

1.3. Simplifies Complex Code
Polymorphism allows you to use a unified interface for different objects, which helps in managing complexity. Rather than managing multiple if-else or switch-case blocks to handle different object types, polymorphism gives you a cleaner, simpler solution.
Example: Handling Multiple Shapes

1. Class Variable
Definition:
A class variable is shared by all instances of a class. It is defined inside the class but outside any methods.
The class variable is bound to the class itself, not the instances, meaning all instances of the class share the same value for that variable.
Access:

Changing the class variable through an instance will affect all instances unless overridden.
Usage:
Class variables are typically used for values or properties that are shared by all instances of a class.

Key Point:
species is a class variable, so it's shared by both dog1 and dog2. When we modify dog1.species, it only changes the instance-level attribute, not the class variable.
2. Instance Variable
Definition:
An instance variable is specific to each instance of a class. It is defined within the __init__() method (or other methods) and is prefixed with self.
Each instance of the class has its own copy of the instance variable, and they can have different values for each object.

Multiple inheritance is a feature in Python where a class can inherit from more than one base class. This allows a derived class to inherit attributes and methods from multiple parent classes, combining functionality from each parent.

When a class inherits from multiple parent classes, it can access all the methods and attributes from all of its parent classes.
```python
#Q(3). Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.?
# Base class Vehicle
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type  # Attribute for vehicle type

    def display_type(self):
        print(f"Vehicle Type: {self.vehicle_type}")

# Derived class Car that inherits from Vehicle
class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)  # Calling the constructor of Vehicle
        self.brand = brand  # Additional attribute for Car

    def display_brand(self):
        print(f"Car Brand: {self.brand}")

# Derived class ElectricCar that inherits from Car
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)  # Calling the constructor of Car
        self.battery_capacity = battery_capacity  # Additional attribute for ElectricCar

    def display_battery(self):
        print(f"Battery Capacity: {self.battery_capacity} kWh")


# Create an instance of ElectricCar
electric_car = ElectricCar("Electric", "Tesla", 75)

# Calling methods from all levels of the inheritance hierarchy
electric_car.display_type()      # Output from Vehicle class
electric_car.display_brand()     # Output from Car class
electric_car.display_battery()   # Output from ElectricCar class
```

```python
# Q(5) Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance?
class BankAccount:
    def __init__(self, initial_balance):
        self._balance = initial_balance  # Private attribute _balance

    # Method to deposit money into the account
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited: ${amount}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money from the account
    def withdraw(self, amount):
        if amount > 0 and amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew: ${amount}")
        elif amount > self._balance:
            print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    # Method to check the current balance
    def check_balance(self):
        print(f"Current Balance: ${self._balance}")

# Create a BankAccount object with an initial balance
account = BankAccount(1000)

# Deposit money into the account
account.deposit(500)

# Withdraw money from the account
account.withdraw(200)

# Check the current balance
account.check_balance()

# Try to withdraw more than available balance
account.withdraw(2000)

# Try to deposit a negative amount
account.deposit(-50)
```

```python
#Q(7). Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.?
class MathOperations:
    # Class method to add two numbers
    @classmethod
    def add_numbers(cls, num1, num2):
        return num1 + num2

    # Static method to subtract two numbers
    @staticmethod
    def subtract_numbers(num1, num2):
        return num1 - num2


# Using the class method and static method
# Calling the class method directly using the class name
sum_result = MathOperations.add_numbers(10, 5)
print(f"Sum: {sum_result}")  # Output: Sum: 15

# Calling the static method directly using the class name
difference_result = MathOperations.subtract_numbers(10, 5)
print(f"Difference: {difference_result}")  # Output: Difference: 5
```

```python
#Q(9) Write a class Fraction with attributes numerator and denominator. Override the str method to display the fraction as "numerator/denominator".?
class Fraction:
    def __init__(self, numerator, denominator):
        # Initializing numerator and denominator attributes
        self.numerator = numerator
        self.denominator = denominator

    # Overriding the __str__ method to represent the fraction as a string
    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

# Creating instances of Fraction
fraction1 = Fraction(3, 4)
fraction2 = Fraction(5, 6)

# Displaying fractions using the __str__ method
print(fraction1)  # Output: 3/4
print(fraction2)  # Output: 5/6
```

```python
#Q(10) Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.?
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator to add two vectors
    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        else:
            raise TypeError("Operand must be of type 'Vector'")

    # Method to display the vector as a string
    def __str__(self):
        return f"({self.x}, {self.y})"

# Creating two vector instances
vector1 = Vector(1, 2)
vector2 = Vector(3, 4)

# Adding the vectors using the overloaded + operator
result = vector1 + vector2

# Printing the result
print(f"Result of vector addition: {result}")  # Output: Result of vector addition: (4, 6)
```

```python
#Q(12) Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades?
class Student:
    def __init__(self, name, grades):
        # Initializing the attributes name and grades
        self.name = name
        self.grades = grades

    def average_grade(self):
        # Method to compute the average of the grades
        if self.grades:
            return sum(self.grades) / len(self.grades)
        else:
            return 0  # If there are no grades, return 0

# Creating an instance of the Student class
student1 = Student("Alice", [85, 90, 78, 92, 88])

# Calling the average_grade method
average = student1.average_grade()
print(f"{student1.name}'s average grade is: {average:.2f}")  # Output: Alice's average grade is: 86.60
```

```python
#Q(13). Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the area.?
class Rectangle:
    def __init__(self):
        # Initializing width and height to None
        self.width = None
        self.height = None

    def set_dimensions(self, width, height):
        # Method to set the dimensions of the rectangle
        self.width = width
        self.height = height

    def area(self):
        # Method to calculate the area of the rectangle
        if self.width is not None and self.height is not None:
            return self.width * self.height
        else:
            return 0  # Return 0 if dimensions are not set yet

# Creating an instance of the Rectangle class
rectangle1 = Rectangle()

# Setting the dimensions using set_dimensions()
rectangle1.set_dimensions(5, 3)

# Calling the area method to calculate the area
area = rectangle1.area()
print(f"The area of the rectangle is: {area}")  # Output: The area of the rectangle is: 15
```

```python
#Q(14). Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.?
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        # Calculate salary based on hours worked and hourly rate
        return self.hours_worked * self.hourly_rate

# Derived class Manager
class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        # Initialize the base class (Employee) and the new attribute bonus
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        # Calculate the salary including the bonus
        base_salary = super().calculate_salary()  # Get the base salary from Employee class
        return base_salary + self.bonus

# Creating an instance of the Employee class
employee1 = Employee("John Doe", 160, 20)  # 160 hours worked, $20 per hour

# Creating an instance of the Manager class
manager1 = Manager("Alice Smith", 160, 30, 5000)  # 160 hours worked, $30 per hour, $5000 bonus

# Calculating and printing salaries
print(f"Employee Salary: ${employee1.calculate_salary()}")  # Output: Employee Salary: $3200
print(f"Manager Salary (including bonus): ${manager1.calculate_salary()}")  # Output: Manager Salary (including bonus): $9500
```

```python
#Q(16)6. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.?
from abc import ABC, abstractmethod

# Abstract class Animal
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass  # Abstract method to be implemented by subclasses

# Derived class Cow
class Cow(Animal):
    def sound(self):
        return "Moo"

# Derived class Sheep
class Sheep(Animal):
    def sound(self):
        return "Baa"

# Creating instances of Cow and Sheep
cow = Cow()
sheep = Sheep()

# Calling the sound method on the instances
print(f"Cow says: {cow.sound()}")  # Output: Cow says: Moo
print(f"Sheep says: {sheep.sound()}")  # Output: Sheep says: Baa
```

```python
#Q(17). Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.?
class Book:
    def __init__(self, title, author, year_published):
        # Initializing the attributes title, author, and year_published
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        # Method to return formatted string with book details
        return f"'{self.title}' by {self.author}, published in {self.year_published}"

# Creating an instance of the Book class
book1 = Book("To Kill a Mockingbird", "Harper Lee", 1960)

# Calling the get_book_info method to get the book's details
book_info = book1.get_book_info()
print(book_info)  # Output: 'To Kill a Mockingbird' by Harper Lee, published in 1960
```

```python
#Q(18)8. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms?
class House:
    def __init__(self, address, price):
        # Initializing the attributes address and price
        self.address = address
        self.price = price

    def get_house_info(self):
        # Method to return house details as a formatted string
        return f"Address: {self.address}, Price: ${self.price}"

# Derived class Mansion
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        # Initializing the base class (House) and the new attribute number_of_rooms
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def get_mansion_info(self):
        # Method to return mansion details including the number of rooms
        return f"{self.get_house_info()}, Number of Rooms: {self.number_of_rooms}"

# Creating an instance of the Mansion class
mansion1 = Mansion("123 Luxury St, Beverly Hills", 5000000, 10)

# Calling the get_mansion_info method to get the mansion's details
mansion_info = mansion1.get_mansion_info()
print(mansion_info)  # Output: Address: 123 Luxury St, Beverly Hills, Price: $5000000, Number of Rooms: 10
```
